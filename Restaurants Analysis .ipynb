# Task 3 --  Price Range Distribution

```python
import matplotlib.pyplot as plt
```

1] Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants ?

```python
Restaurants = Data['Restaurant Name'].value_counts()
Restaurants
```

Showing The Bar-Graph

```python
Price_range = list()
for i in Restaurants.keys():
    Res = Data[Data['Restaurant Name'] == i]
    Price = Res['Price range'].max()
    Price_range.append(Price)
```

```python
x = list(Restaurants.keys()[0:10])
y = Price_range[0:10]
plt.figure(figsize = (10,4))
plt.bar(x,y,width = 0.4 )
plt.xlabel('Restaurants')
plt.ylabel('Reice range')
plt.xticks(rotation=30)
plt.title("Restaurants by Price range",fontsize=13)
plt.show()
```

2] Calculate the percentage of restaurants in each price range category ?

```python
Rest = Data["Restaurant Name"].count()
price_range = Data["Price range"].value_counts()
per_restaurants = (price_range/Rest)*100
per_restaurants
```

Showing The Pie-Chart

```python
plt.pie(price_range,labels = price_range.keys(),autopct='%1.2f%%')
plt.show()
```

2] Calculate the average number of votes received by restaurants ?

```python
Restaurants = Data['Restaurant Name'].value_counts()
rest = list()
num_v = list()
for i in Restaurants.keys():
    Rest = Data[Data['Restaurant Name']== i]
    Num_votes = np.average(Rest["Votes"]).round(2)
    rest.append(i)
    num_v.append(Num_votes)
    #print(i,'---> average votes :- ',Num_votes)
votses_by_rest = pd.DataFrame(list(zip(rest,num_v)),columns = ["Restaurants","Votes"])
votses_by_rest.sort_values('Votes',ascending = False)
```
