1] Analyze the distribution of aggregate ratings and determine the most common rating range ?

```python
sns.lineplot(data=rating_df,x = 'Agg_rating',y = 'count_of_rating')
```

2] Calculate the average number of votes received by restaurants ?

```python
restaurants = restaurant_df['Restaurant Name'].value_counts()
rest = list()
average_votes = list()
for i in restaurants.keys():
    Rest = restaurant_df[restaurant_df['Restaurant Name']== i]
    avg_votes = np.average(Rest["Votes"]).round(2)
    rest.append(i)
    average_votes.append(avg_votes)
votes_by_rest = pd.DataFrame(list(zip(rest,average_votes)),columns = ["restaurants","votes"])
votes_by_rest.sort_values('votes',ascending = False,inplace=True)
votes_by_rest
```

```python
#Barplot
x = votes_by_rest['restaurants'][0:15]
y = votes_by_rest['votes'][0:15]
plt.bar(x,y)
plt.title('average number of votes received by restaurants')
plt.xticks(rotation = 90)
plt.xlabel('restaurants')
plt.ylabel('average votes')
plt.show()
```

```python
Rating_text = list()
Agg_max_rate = list()
Agg_min_rate = list()
for i in Text.keys():
    Rate_text = restaurant_df[restaurant_df["Rating text"] == i]
    Agg_max_rating = np.max(Rate_text["Aggregate rating"])
    Agg_min_rating = np.min(Rate_text["Aggregate rating"])
    Rating_text.append(i)
    Agg_max_rate.append(Agg_max_rating)
    Agg_min_rate.append(Agg_min_rating)
review_len_rating = pd.DataFrame(list(zip(Rating_text,Agg_max_rate,Agg_min_rate)),columns = ["Rating text","Max rating","Min rating"])
review_len_rating
```
